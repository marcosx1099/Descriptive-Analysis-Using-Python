```python
import pandas as pd
import numpy as np
import statistics as st
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
from scipy.stats import kurtosis
```

```python
#task 1 drop all missing values in columns i included below
df = pd.read_csv('week5pythonDS-ChallengeFile.csv')
df.dropna(subset = ['Cylinders',"Kerb_Weight",'Make','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque'],inplace = True)
```

```python
#task2 # replacing all occurances of kg with "" (nothing) i.e removing "kg"

df['Kerb_Weight'] = df['Kerb_Weight'].str.replace(r" kg",'') 
df['Kerb_Weight'] = df['Kerb_Weight'].str.strip() # stripping white spaces if any
df['Kerb_Weight'] = df['Kerb_Weight'].astype(float)
df[['Kerb_Weight']]
```

```python
#cleaning the power column  
df['Power'] = df['Power'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*",'', regex=True) 
df["Power"] = df["Power"].str.strip() # stripping white spaces if any
df['Power'] = df['Power'].astype(float)
df['Power']
```

```python
#cleaning the torque column 


df['Torque'] = df['Torque'].str.replace(r"-",'', regex=True) # Removing any occurances of "-" in this the column
df['Torque'] = df['Torque'].str.replace(r"[a-zA-Z]*@[0-9]*[a-zA-Z]*",'', regex=True) 
df['Torque'] = df["Torque"].str.strip() # stripping white spaces if any 
df['Torque'] = df['Torque'].astype(float)
df['Torque']

df[['Kerb_Weight','Power','Torque']]
```
